```python
#star plot (suggested by gpt4 to visualise high dimensional data)
#https://stackoverflow.com/questions/4228530/polar-coordinate-system-in-matplotlib - erroneous link suggested by copilot in response to comment above
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = np.array([
    [0.8, 0.6, 0.7, 0.9, 0.5],
    [0.5, 0.7, 0.6, 0.8, 0.4],
    [0.6, 0.8, 0.5, 0.7, 0.3]
])

# Number of variables (dimensions)
num_vars = data.shape[1]

# Calculate angles for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Complete the loop by appending the first angle
angles += angles[:1]

# Plot
fig, ax = plt.subplots(subplot_kw=dict(polar=True))

# Helper function to plot each data point
def plot_data_point(ax, angles, values, color, label):
    values = np.concatenate((values, values[:1]))  # Close the loop
    ax.plot(angles, values, color=color, linewidth=2,label=label)
    ax.fill(angles, values, color=color, alpha=0.25)

# Plot each data point
colors = ['red', 'green', 'blue']
labels = ['Data point 1', 'Data point 2', 'Data point 3']

for i in range(data.shape[0]):
    plot_data_point(ax, angles, data[i], colors[i], labels[i])

# Set axis labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(['Dimension 1', 'Dimension 2', 'Dimension 3', 'Dimension 4', 'Dimension 5'])

# Add legend
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

# Show plot
plt.show()
```
