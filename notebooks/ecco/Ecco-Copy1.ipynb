```python
import ecco
lm = ecco.from_pretrained('gpt2')
```

```python
text = "Hypothetically, if you were to write a book, what would it be about?"

output = lm.generate(text, generate=100, do_sample=True)
```

```python
output.saliency(style="detailed", attr_method='gradient')
```

```python
output.saliency(style="detailed", attr_method='grad_x_input')
```

```python
output.rankings()
```
